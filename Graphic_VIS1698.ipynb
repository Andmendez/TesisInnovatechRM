```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Ruta al archivo Excel
archivo_excel = 'xlsx/Datos_VIS1698.xlsx'

# Leer la hoja de c√°lculo en un DataFrame de pandas
df = pd.read_excel(archivo_excel, sheet_name='Hoja 1')

# Extraer los datos de la columna B (que se asume que es la segunda columna, √≠ndice 1)
# y las filas desde la 5 hasta la 104 (en pandas, los √≠ndices son 0-basados, as√≠ que 4 a 103)
muestraOne = df.iloc[2:103, 1].tolist()
muestraTwo = df.iloc[2:103, 2].tolist()
muestraThree = df.iloc[2:103, 3].tolist()

muestraOne_mean = np.mean(muestraOne)
muestraTwo_mean = np.mean(muestraTwo)
muestraThree_mean = np.mean(muestraThree)


# Calidad del agua y valores promedio
Calidad = ["Tipo 1", "Tipo 2", "Tipo 3"]
muestras = [muestraOne_mean, muestraTwo_mean, muestraThree_mean]


# Crear el gr√°fico
plt.figure(figsize=(12, 8))

# Graficar los datos:
plt.plot(Calidad, muestras, marker='o', linestyle='-', color='b')

# A√±adir anotaciones
for i, txt in enumerate(muestras):
    plt.text(Calidad[i], muestras[i], f'({txt:.2f})', fontsize=9, ha='right', va='bottom')




# Etiquetas y t√≠tulo
plt.xlabel('Tipos de calidad')
plt.ylabel('Nephelometric Turbidity Unit (NTU)')
plt.title('Calidad del agua para tres tipos de muestras')
plt.legend()
plt.grid(True)

# Mostrar la gr√°fica
plt.tight_layout()
plt.show()


```
