```python
import pandas as pd

# Ruta al archivo Excel
archivo_excel = 'xlsx/Datos_NPK.xlsx'

# Leer la hoja de cálculo en un DataFrame de pandas
df = pd.read_excel(archivo_excel, sheet_name='Datos NPK')

# Extraer los datos de la columna B (que se asume que es la segunda columna, índice 1)
# y las filas desde la 5 hasta la 104 (en pandas, los índices son 0-basados, así que 4 a 103)
N_10 = df.iloc[3:103, 1].tolist()
P_10 = df.iloc[3:103, 2].tolist()
K_10 = df.iloc[3:103, 3].tolist()

N_30 = df.iloc[3:103, 5].tolist()
P_30 = df.iloc[3:103, 6].tolist()
K_30 = df.iloc[3:103, 7].tolist()

N_50 = df.iloc[3:103, 9].tolist()
P_50 = df.iloc[3:103, 10].tolist()
K_50 = df.iloc[3:103, 11].tolist()

N_10_H = df.iloc[110:210, 1].tolist()
P_10_H = df.iloc[110:210, 2].tolist()
K_10_H = df.iloc[110:210, 3].tolist()

N_30_H = df.iloc[110:210, 5].tolist()
P_30_H = df.iloc[110:210, 6].tolist()
K_30_H = df.iloc[110:210, 7].tolist()

N_50_H = df.iloc[110:210, 9].tolist()
P_50_H = df.iloc[110:210, 10].tolist()
K_50_H = df.iloc[110:210, 11].tolist()

# Mostrar los datos extraídos
print(N_10)
print(len(P_10))
print(len(K_10))

print(len(N_30))
print(len(P_30))
print(len(K_30))

print(N_50)
print(len(P_50))
print(len(K_50))

print(N_10_H)

```

```python
import matplotlib.pyplot as plt
import numpy as np



# Calcular el promedio para cada altura
N_10_mean = np.mean(N_10)
N_30_mean = np.mean(N_30)
N_50_mean = np.mean(N_50)

N_10_H_mean = np.mean(N_10_H)
N_30_H_mean = np.mean(N_30_H)
N_50_H_mean = np.mean(N_50_H)


# Alturas y valores promedio
alturas = [10, 30, 50]
niveles_nitrogeno = [N_10_mean, N_30_mean, N_50_mean]
niveles_nitrogeno_H = [N_10_H_mean, N_30_H_mean, N_50_H_mean]

# Crear el gráfico
plt.figure(figsize=(12, 8))

# Graficar los datos para baja humedad
plt.plot(alturas, niveles_nitrogeno, marker='o', linestyle='-', color='b', label='Baja Humedad')


# Añadir anotaciones
for i, txt in enumerate(niveles_nitrogeno):
    plt.text(alturas[i], niveles_nitrogeno[i], f'({txt:.2f})', fontsize=9, ha='right', va='bottom')



# Graficar los datos para alta humedad
plt.plot(alturas, niveles_nitrogeno_H, marker='o', linestyle='--', color='r', label='Alta Humedad')


# Añadir anotaciones
for i, txt in enumerate(niveles_nitrogeno_H):
    plt.text(alturas[i], niveles_nitrogeno_H[i], f'({txt:.2f})', fontsize=9, ha='right', va='bottom')


# Etiquetas y título
plt.xlabel('Altura (cm)')
plt.ylabel('Concentración de Nitrógeno (mg/kg)')
plt.title('Nivel de Nitrógeno vs Altura segun Humedad del suelo')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```
